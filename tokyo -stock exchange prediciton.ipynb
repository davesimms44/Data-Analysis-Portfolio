# using msno


```python
import missingno as msno
```

```python
#Bar chart
msno.bar(stocks)
```

```python
#Matrix Chart      
#great for time series data 
msno.matrix(stocks)
```

```python
#Heatmap 
#used to identify correlations of the nullity between each of the different columns ,
 # is there a relationship between the missinness of one variable and another?
msno.heatmap(stocks)
```

```python
#dendogram
#groups together columns with strong correlation of nullity
#if a number of columns are grouped together at level 0 than the presence of nulls is one of those columns is
#directly related to the presence or absence of nulls in the other columns
#the more seperated the columns in the tree the less likely the null values can be correlated between the columns
msno.dendrogram(stocks)
```
